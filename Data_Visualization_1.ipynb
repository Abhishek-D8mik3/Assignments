```python
#Q8 Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(42)
size = 100


feature1 = np.random.randn(size)
feature2 = 0.8 * feature1 + np.random.randn(size) * 0.2
feature3 = 0.5 * feature1 - 0.3 * feature2 + np.random.randn(size) * 0.5


df = pd.DataFrame({
    'Feature1': feature1,
    'Feature2': feature2,
    'Feature3': feature3
})

corr_matrix = df.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)


plt.title('Correlation Matrix Heatmap')


plt.show()
```

```python
#Q9 Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

import plotly.graph_objs as go
np.random.seed(30)
data = { 'X': np.random.uniform(-10, 10, 300),
         'Y': np.random.uniform(-10, 10, 300),
         'Z': np.random.uniform(-10, 10, 300) }
df = pd.DataFrame(data)

trace = go.Scatter3d(
    x=df['X'],  y=df['Y'],  z=df['Z'],  mode='markers', marker=dict(size=5,  color=df['Z'],  colorscale='Viridis',  opacity=0.8 ))

layout = go.Layout(
    title='3D Scatter Plot of Synthetic Data',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

fig = go.Figure(data=[trace], layout=layout)

fig.show()

```

```python
#Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

np.random.seed(20)
data = { 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
        'Day': np.random.choice(range(1, 31), 100),
         'Sales': np.random.randint(100, 500, 100) }
df = pd.DataFrame(data)

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(100, 500, 100)
}
df = pd.DataFrame(data)

sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, annot=True, cmap='YlGnBu', linewidths=0.5)

plt.title('Sales Variation Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

plt.show()
```

```python
#Q12  Using the given x and y data, generate a 3D surface plot to visualize the function  z = sin(np.sqrt(x**2 + y**2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = { 'X': x.flatten(), 'Y': y.flatten(), 'Z': z.flatten() }
df = pd.DataFrame(data)


data = {'X': x.flatten(), 'Y': y.flatten(), 'Z': z.flatten()}
df = pd.DataFrame(data)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')


ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

plt.show()
```

```python
#Q13  Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

np.random.seed(25)
data = { 'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
        'Population': np.random.randint(100, 1000, 5),
        'GDP': np.random.randint(500, 2000, 5) }
df = pd.DataFrame(data)

fig = px.scatter(df, x='GDP', y='Population', size='Population', text='Country',
                 title='Bubble Chart: Population vs GDP',
                 labels={'GDP': 'GDP (in billions)', 'Population': 'Population (in millions)'})


fig.update_traces(textposition='top center')
fig.show()
```

```python
#Q14 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

x = np.linspace(0, 10, 100)
y = np.sin(x)


output_notebook()
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y', width=800, height=400)
p.line(x, y, legend_label='y = sin(x)', line_width=2, color='blue')


show(p)
```

```python
#Q15 .Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

from bokeh.models import ColumnDataSource


np.random.seed(42)
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100
sizes = np.random.randint(10, 100, size=50)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=50)


source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})


output_notebook()
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)


p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)


show(p)
```
