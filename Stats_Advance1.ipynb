```python
#Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25,
# calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?
#Task: Write Python code to calculate the z-score and p-value for the given data. Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.

import scipy.stats as stats
import math

# Given data
sample_mean = 105
population_mean = 100
std_deviation = 15
sample_size = 25
alpha = 0.05  # significance level

# Calculate the z-score
z_score = (sample_mean - population_mean) / (std_deviation / math.sqrt(sample_size))

# Calculate the p-value from the z-score
p_value = stats.norm.cdf(z_score)

# Print the z-score and p-value
print("Z-Score:", z_score)
print("P-Value:", p_value)

# Hypothesis testing decision
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python.
#Generate 1,000 samples and plot the distribution. What is the expected mean and variance?
#Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.
#Objective: Understand the properties of a binomial distribution and verify them through simulation.

import numpy as np
import matplotlib.pyplot as plt

# Given parameters
n_trials = 10
prob_success = 0.6
n_samples = 1000

# Generate 1000 samples of binomial distribution
samples = np.random.binomial(n_trials, prob_success, n_samples)

# Plot the distribution
plt.hist(samples, bins=np.arange(-0.5, n_trials+1.5, 1), edgecolor='black', alpha=0.7)
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n_trials+1))
plt.show()

# Calculate the mean and variance of the generated samples
mean = np.mean(samples)
variance = np.var(samples)

print(f"Expected Mean: {mean}")
print(f"Expected Variance: {variance}")
```
